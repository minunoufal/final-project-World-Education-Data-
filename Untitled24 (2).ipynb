```python
sns.scatterplot(x='gov_exp_pct_gdp',y='lit_rate_adult_pct',data=df)
plt.title("GDP Vs Literacy Rate")
plt.show()
```

```python
df.groupby('country')['lit_rate_adult_pct'].mean()
```

```python
pd.pivot_table(df,values='lit_rate_adult_pct',index='country',aggfunc='mean')
```

```python
corr=df.select_dtypes(include='number').corr()
sns.heatmap(corr,annot=True)
plt.title("Corelation Analysis")
plt.show()
```

```python
Visualization
```

```python
df.groupby('country')['lit_rate_adult_pct'].mean().plot(kind='bar')
plt.title("Literacy Rate by country")
plt.xlabel("Country")
plt.ylabel("Literacy Rate")
plt.show()
```
